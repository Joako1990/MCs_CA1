### Another visualization showing percentages

```python
def pltpie(year,title,dataFrame,chart):  #Creates a function to plot Pie chart based on the year
    
    #Create df
    pop=dataFrame[dataFrame.Year==year]
    
    
    #Create variables
    labels=pop.Nationality
    sizes=pop.VALUE
    
    if chart == 'Sex':
        
        #Plot Pie Chart
        fig, ax=plt.subplots(figsize=(10,5))
        ax.pie(sizes,labels=labels,autopct='%1.1f%%',startangle=90);
        ax.axis('equal')
        ax.set_title(title + str(year))
        plt.legend(dataFrame.Sex,loc="best")
        
    elif chart == 'Nationality':
        
        #Plot Pie Chart
        fig, ax=plt.subplots(figsize=(10,5))
        ax.pie(sizes,labels=labels,autopct='%1.1f%%',startangle=90);
        ax.axis('equal')
        ax.set_title(title + str(year))
        plt.legend(dataFrame.Nationality,loc="best")
        
    elif chart == 'Year':

        #Plot Pie Chart
        fig, ax=plt.subplots(figsize=(10,5))
        ax.pie(sizes,labels=labels,autopct='%1.1f%%',startangle=90);
        ax.axis('equal')
        ax.set_title(title + str(year))
        plt.legend(dataFrame.Year,loc="best")
        
```

## Binomial Bistribution

```python
# Import bimomial librarie

from scipy.stats import binom
import math
```

#### First define the formula for binomial distribution

$$f(x,n,p)={n\choose x} p^x (1-p)^{n-x}$$

where: $${n \choose x} = \frac{n!}{x!(n-x)!}$$

and:
  
    *x= Number of people we want to find
    *n= Sample of the group we are selecting
    *p= Probability of the people to be of a certain nationality
    

```python
#According to our dataset, we can calculate the probability of finding x number of people 
#from a specific group of nationality in Ireland based on the year by knowing the % of people from that group, which we can
#ger from our pie chart

def binomialD(x,n,p,nat):
    p=p/100
    a= math.factorial(n) / (math.factorial(x) * math.factorial(n-x))
    b= p**x * (1-p)**(n-x)
    res=a*b
    #return a*b
    #return'Probability of finding ' + str(x), 'persons from ' + nat, 'in a group of ' + str(n), 'is ' + f"{res:.2%}"
    print('Probability of finding ' + str(x), 'persons from ' + nat, 'in a group of ' + str(n), 'is ' + f"{res:.6%}")
```

```python
def binomial(x,n,p):
    p=p/100
    a= math.factorial(n) / (math.factorial(x) * math.factorial(n-x))
    b= p**x * (1-p)**(n-x)
    return a*b
```

```python
#We can see 88.2% of the population is Irish so:
#We are going to find the propapility of finding an x number irish people in a sample of n
binomialD(7,15,88.2,'Ireland')
```

```python
#Similarly we can find the propability of finding a person from a different nationality but it'll be a tiny number due
#to the difference between irish and other nationalities population
binomialD(7,15,5.6,'ther nationalities')
```

```python
#Similarly we can find the propability of finding a person from the UK but it'll be a tiny number due
#to the difference between irish and UK population
binomialD(7,15,2.7,'UK')
```

## Bernoulli Distribution

```python
# We can find the probability of one event of two to occure using the Bernoulli distribution, in this case, the 
# probability that a person is Male or Female
```

```python
#First obtain probabilities of a person to be of a specific gender each year where 0=Male & 1=Female 
nc_df.reset_index(drop=True)
```

```python
pop_df=male_df.groupby('Year').sum('VALUE')
pop_df
```

```python
pop_df.insert(0, "Year", [2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020,2021,2022], True)
pop_df.reset_index(drop=True)
```

```python
pop_df['Sex']='M'
pop_df
```

```python
female_pop=female_df.groupby('Year').sum('VALUE')
female_pop
```

```python
female_pop.insert(0, "Year", [2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020,2021,2022], True)
female_pop.reset_index(drop=True)
```

```python
female_pop['Sex']='F'
female_pop
```

```python
pop_df['Total']=pop_df['VALUE']+female_pop['VALUE']
pop_df
```

```python
female_pop['Total']=pop_df['Total']
female_pop
```

```python
pop_df['Probability']=pop_df['VALUE'] / pop_df['Total']
pop_df
```

```python
female_pop['Probability']=female_pop['VALUE'] / pop_df['Total']
female_pop
```

```python
pop1_df=pop_df.append(female_pop)
pop1_df=pop1_df.reset_index(drop=True)
pop1_df
```

```python
pop1_df = pop1_df.sort_values('Year', ascending=True)
pop1_df.reset_index(drop=True,inplace=True)
```

```python
pop1_df
```

```python
#Now lets plot the distribution

plt.figure(figsize=(10,5))
sns.barplot(x=pop1_df['Sex'], y=pop1_df['Probability'], data=pop1_df)
plt.title('Bernoulli Distribution');
#plt.legend(bbox_to_anchor=(1.15,1), loc='upper right');
```

```python
#Now lets plot the distribution

plt.figure(figsize=(10,5))
sns.barplot(x=pop1_df['Sex'], y=pop1_df['Probability'], hue='Year', data=pop1_df)
plt.title('Bernoulli Distribution');
plt.legend(bbox_to_anchor=(1.15,1), loc='upper right');
```
