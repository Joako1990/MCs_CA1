### Another visualization showing percentages

```python
def pltpie(year,title,dataFrame,chart):  #Creates a function to plot Pie chart based on the year
    
    #Create df
    pop=dataFrame[dataFrame.Year==year]
    
    
    #Create variables
    labels=pop.Nationality
    sizes=pop.VALUE
    
    if chart == 'Sex':
        
        #Plot Pie Chart
        fig, ax=plt.subplots(figsize=(10,5))
        ax.pie(sizes,labels=labels,autopct='%1.1f%%',startangle=90);
        ax.axis('equal')
        ax.set_title(title + str(year))
        plt.legend(dataFrame.Sex,loc="best")
        
    elif chart == 'Nationality':
        
        #Plot Pie Chart
        fig, ax=plt.subplots(figsize=(10,5))
        ax.pie(sizes,labels=labels,autopct='%1.1f%%',startangle=90);
        ax.axis('equal')
        ax.set_title(title + str(year))
        plt.legend(dataFrame.Nationality,loc="best")
        
    elif chart == 'Year':

        #Plot Pie Chart
        fig, ax=plt.subplots(figsize=(10,5))
        ax.pie(sizes,labels=labels,autopct='%1.1f%%',startangle=90);
        ax.axis('equal')
        ax.set_title(title + str(year))
        plt.legend(dataFrame.Year,loc="best")
```

## Binomial Bistribution

```python
# Import bimomial librarie

from scipy.stats import binom
import math
```

#### First define the formula for binomial distribution

$$f(x,n,p)={n\choose x} p^x (1-p)^{n-x}$$

where: $${n \choose x} = \frac{n!}{x!(n-x)!}$$

and:
  
    *x= Number of people we want to find
    *n= Sample of the group we are selecting
    *p= Probability of the people to be of a certain nationality
    

```python
#According to our dataset, we can calculate the probability of finding x number of people 
#from a specific group of nationality in Ireland based on the year by knowing the % of people from that group, which we can
#ger from our pie chart

def binomialD(x,n,p,nat):
    p=p/100
    a= math.factorial(n) / (math.factorial(x) * math.factorial(n-x))
    b= p**x * (1-p)**(n-x)
    res=a*b
    #return a*b
    #return'Probability of finding ' + str(x), 'persons from ' + nat, 'in a group of ' + str(n), 'is ' + f"{res:.2%}"
    print('Probability of finding ' + str(x), 'persons from ' + nat, 'in a group of ' + str(n), 'is ' + f"{res:.6%}")
```

```python
#We can see 88.2% of the population is Irish so:
#We are going to find the propapility of finding an x number irish people in a sample of n
binomialD(7,15,88.2,'Ireland')
```

```python
#Similarly we can find the propability of finding a person from a different nationality but it'll be a tiny number due
#to the difference between irish and other nationalities population
binomialD(7,15,5.6,'ther nationalities')
```
